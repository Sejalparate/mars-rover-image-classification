## Pre-processing

```python
# Modify image path names within a dataset
def new_path_function(jpg_path, jpg_labels, new_jpg_list, new_label_list, splitting_string = "calibrated"):
    for image_path, path_label in zip(jpg_path, jpg_labels):
        ID_pathing, exporting_type = image_path.split(splitting_string)
        new_file_path_name = "Mars Surface and Curiosity Image dataset/images" + str(exporting_type)
        new_jpg_list.append(new_file_path_name)
        new_label_list.append(path_label)
```

```python
new_jpg_path_train = []
new_labels_train = []
splitting_string = "calibrated"
new_path_function(train_csv.JPG, train_csv.LABELS, new_jpg_path_train, new_labels_train, splitting_string)
```

```python
new_jpg_path_test = []
new_labels_test = []
splitting_string = "calibrated"
new_path_function(test_csv.JPG, test_csv.LABELS, new_jpg_path_test, new_labels_test, splitting_string)
```

```python
new_jpg_path_val = []
new_labels_val = []
splitting_string = "calibrated"
new_path_function(val_csv.JPG, val_csv.LABELS, new_jpg_path_val, new_labels_val, splitting_string)
```

```python
# Create dataframe
train_jpg_series = pd.Series(new_jpg_path_train, name = 'JPG').astype(str)
train_labels_series = pd.Series(new_labels_train, name = 'CATEGORY')

main_train_data = pd.concat([train_jpg_series, train_labels_series], axis = 1)
main_train_data.head(-1)
```

```python
test_jpg_series = pd.Series(new_jpg_path_test, name = 'JPG').astype(str)
test_labels_series = pd.Series(new_labels_test, name = 'CATEGORY')

main_test_data = pd.concat([test_jpg_series, test_labels_series], axis = 1)
main_test_data.head(-1)
```

```python
val_jpg_series = pd.Series(new_jpg_path_val, name = 'JPG').astype(str)
val_labels_series = pd.Series(new_labels_val, name = 'CATEGORY')

main_val_data = pd.concat([val_jpg_series, val_labels_series], axis = 1)
main_val_data.head(-1)
```

```python
# Merging and shuffling
frame_list = [main_train_data, main_test_data, main_val_data]
main_mars_data = pd.concat(frame_list)
main_mars_data = main_mars_data.sample(frac = 1).reset_index(drop = True)
main_mars_data.head(-1)
```

## Image processing

```python
def vision_rgb(image_path):
    pick_img = image_path
    pick_img = cv2.cvtColor(cv2.imread(pick_img), cv2.COLOR_BGR2RGB)

    plt.xlabel(pick_img.shape)
    plt.ylabel(pick_img.size)
    plt.imshow(pick_img);
```

```python
def vision_bgr(image_path):
    plt.xlabel(image_path.shape)
    plt.ylabel(image_path.size)
    plt.imshow(image_path);
```

```python
def threshold(image_path):
    pick_img = image_path
    pick_img = cv2.cvtColor(cv2.imread(pick_img), cv2.COLOR_BGR2RGB)
    _, threshold_img = cv2.threshold(pick_img, 200, 255, cv2.THRESH_BINARY_INV)

    plt.xlabel(threshold_img.shape)
    plt.ylabel(threshold_img.size)
    plt.imshow(threshold_img);
```

```python
def canny(image_path):
    canny_image = cv2.Canny(image_path, 10, 100)

    plt.xlabel(canny_image.shape)
    plt.ylabel(canny_image.size)
    plt.imshow(canny_image);
```

```python
def drawing_contours(image_path):
    canny_img = cv2.Canny(image_path, 10, 100)
    contour, _ = cv2.findContours(canny_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    drawing_contour = cv2.drawContours(image_path, contour, -1, (255, 0, 0), 2)

    plt.xlabel(drawing_contour.shape)
    plt.ylabel(drawing_contour.size)
    plt.imshow(drawing_contour);
```

```python
transformed_Y_train = []
transformed_Y_train_labels = []

for Y_image, Y_labels in zip(X_train.JPG, X_train.CATEGORY):
    Y_image = cv2.cvtColor(cv2.imread(Y_image), cv2.COLOR_BGR2RGB)
    Y_image = cv2.resize(Y_image, (180, 180))
    Y_image = Y_image / 255.0
    transformed_Y_train.append(Y_image)
    transformed_Y_train_labels.append(Y_labels)
```

```python
y_s_train_img = np.array(transformed_Y_train)
y_s_train_labels = to_categorical(transformed_Y_train_labels)
```

```python
print(y_s_train_img.shape)
print(y_s_train_labels.shape)
print(x_ae_train.shape)
```

## Model Architecture

```python
Early_Stopper = tf.keras.callbacks.EarlyStopping(monitor = "loss", patience = 3, mode = "min")
Checkpoint_Model = tf.keras.callbacks.ModelCheckpoint(monitor = "val_accuracy", save_best_only = True, save_weights_only = True,
                                                      filepath = "./modelcheck")
Reduce_Model = tf.keras.callbacks.ReduceLROnPlateau(monitor = "accuracy", factor = 0.1, patience = 7)
```

```python
Input_Layer = tf.keras.Input(shape = (180, 180, 3))

x = Conv2D(32, (3, 3), activation = "relu", padding = "same")(Input_Layer)
x = MaxPooling2D((2, 2))(x)
x = Conv2D(64,(3, 3), activation = "relu", padding = "same")(x)
x = MaxPooling2D((2, 2))(x)
x = Conv2D(128, (2, 2), activation = "relu", padding = "same")(x)
x = MaxPooling2D((2, 2))(x)
x = Conv2D(256, (2, 2), activation = "relu", padding = "same")(x)
x = GlobalMaxPooling2D()(x)
x = Dense(128, activation = "relu")(x)
x = Dropout(0.5)(x)

class_prediction_layer = Dense(25, activation = "softmax",name = "CLASS_PREDICTION")(x)

encoder = Dense(128, activation = "relu")(Input_Layer)
encoder = Dense(64, activation = "relu")(encoder)
encoder = Dense(32, activation = "relu")(encoder)

# Reconstruct the original input image from the compressed representation
decoder = Dense(64, input_shape = [32], activation = "relu")(encoder)
decoder = Dense(128, activation = "relu")(decoder)
ae_output = Dense(3, activation = "sigmoid", name = "AE_OUTPUT")(decoder)
```

```python
configure_model = Model(Input_Layer, [class_prediction_layer, ae_output])
```

```python
print(configure_model.summary())
```

```python
configure_model.compile(optimizer = "adam", loss = {"CLASS_PREDICTION": "categorical_crossentropy", "AE_OUTPUT": "binary_crossentropy"},
                        metrics = ["accuracy"])
```

```python
model_configure_total = configure_model.fit(y_s_train_img, [y_s_train_labels, x_ae_train], epochs = 50, batch_size = 32,
                    callbacks = [Early_Stopper, Checkpoint_Model, Reduce_Model])
```

```python
# Create mapping from class indices to class name
class_indices = {}
with open("Mars Surface and Curiosity Image dataset/Labels Information.txt", "r") as f:
    for line in f:
        index, class_name = line.strip().split(maxsplit = 1)  # Split at most once for tabs
        class_indices[int(index)] = class_name.strip()  # Convert index to integer and strip whitespace
for index, class_name in class_indices.items():
    print(f"{index}: {class_name}")
```
